```python
# Leemos nuestros dataframes
movies = pd.read_csv("./ml-latest-small/movies.csv", sep=",")
ratings = pd.read_csv("./ml-latest-small/ratings.csv", sep=",")
tags = pd.read_csv("./ml-latest-small/tags.csv", sep=",")
links = pd.read_csv("./ml-latest-small/links.csv", sep=",")

# Eliminamos nulos
movies.dropna(inplace=True)
ratings.dropna(inplace=True)
tags.dropna(inplace=True)
links.dropna(inplace=True)
```

```python
# No queremos trabajar con timestamp
ratings = ratings.drop('timestamp', 1)
```

```python
ratings.groupby(["rating"])["userId"].count()
```

```python
idEspecifico = 1
nuestroUser = ratings.loc[ratings['userId'] == idEspecifico]
nuestroUser
```

```python
df_matrix = pd.pivot_table(ratings, values='rating', index='userId', columns='movieId').fillna(0)
df_matrix
```

```python
dfs = np.array_split(df_matrix, 4)
df1 = pd.DataFrame(dfs[0])
df2 = pd.DataFrame(dfs[1])
df3 = pd.DataFrame(dfs[2])
df4 = pd.DataFrame(dfs[3])
del df_matrix, dfs

sim_matrix1 = 1 - sklearn.metrics.pairwise.cosine_distances(df1)
np.savetxt("./ml-latest-small/sim_matrix1.csv", sim_matrix1, delimiter=",")
del sim_matrix1
sim_matrix2 = 1 - sklearn.metrics.pairwise.cosine_distances(df2)
np.savetxt("./ml-latest-small/sim_matrix2.csv", sim_matrix2, delimiter=",")
del sim_matrix2
sim_matrix3 = 1 - sklearn.metrics.pairwise.cosine_distances(df3)
np.savetxt("./ml-latest-small/sim_matrix3.csv", sim_matrix3, delimiter=",")
del sim_matrix3
sim_matrix4 = 1 - sklearn.metrics.pairwise.cosine_distances(df4)
np.savetxt("./ml-latest-small/sim_matrix4.csv", sim_matrix4, delimiter=",")
del sim_matrix4
```
